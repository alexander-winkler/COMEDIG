```python
from lxml import etree
from glob import glob
import csv
import shutil
```

```python
xml_files = glob('xml/*.xml')
```

```python
project = []
with open('projekte_utf.tsv.csv','r', encoding="utf-8") as IN:
    reader = csv.DictReader(IN, delimiter=',')
    for r in reader:
        project.append(r)
```

```python
images = {}
for p in project:
    for img in glob("TRASCRIZIONI _ TRADUZIONI/" + p.get('img_folder') + "/**/*", recursive = True ):
        file_name = img.split('/')[-1]
        if file_name.endswith(('jpeg','jpg','tiff')):
            images[file_name] = img
        
```

```python
# Ersetzungsmatrix f√ºr facs-Ids
graphic_id = {}
```

```python
img_refs = []
for p in project:
    xml = f"legacy_xml/{p.get('XML')}"
    xml_target = f"xml/{p.get('proj_id')}.xml"
    print(f"Drama:\t{p.get('proj_id')}")
    print(f"FROM {xml} TO {xml_target}")
    tree = etree.parse(xml)
    for pb in tree.findall('.//{*}pb'):
        print(pb.get('facs'))
        try:
            pb.set('facs', "#" + graphic_id.get(pb.get('facs')[1:]))
        except:
            print(pb.get('facs'),graphic_id.get(pb.get('facs')[1:]))
            break
    surface = tree.findall('.//{*}facsimile/{*}surface')
    for s in surface:
        lan = s.attrib.get('type')
        print(f"\tSprache: {lan}")
        for n,g in enumerate(s.findall('./{*}graphic'), start = 1):
            img_url = g.attrib.get('url')
            img_file = img_url.split('/')[-1]
            img_refs.append(img_file)
            gr_id = g.attrib.get('{http://www.w3.org/XML/1998/namespace}id')
            try:
                img_path = images.get(img_file)
                suffix = img_path.split('.')[-1]
                target_id = f"{p.get('proj_id')}_{lan}_{n:03d}"
                graphic_id[gr_id] = target_id
                img_target = f"/img/{target_id}.{suffix}"
                print(f"\t\tGraphic id:\t{gr_id}\tImage URL:{img_url}")
                g.set('{http://www.w3.org/XML/1998/namespace}id', graphic_id.get(gr_id))
                g.set('url',img_target)
                img_url = img_target
                print(f"\t\tIMGAGE FROM\t{img_path}\tTO\t{img_target}")
                #shutil.copyfile(img_path,f"./{img_target}")
            except Exception as e:
                print(img_url, img_path)
                print(e)
    with open(xml_target, 'w') as OUT:
        output = etree.tostring(tree, encoding='unicode')
        OUT.write(output)
            
```
